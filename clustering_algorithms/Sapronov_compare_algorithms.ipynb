```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.cluster import DBSCAN
```

```python
%matplotlib inline
```

```python
random_state = 170
```

```python
def load_dataset(fn):
    dataset = []
    with open(fn, 'r') as dsf:
        for line in dsf:
            point = line.strip().split('\t')
            dataset.append([float(point[0]), float(point[1])])
    return dataset
```

```python
dataset = load_dataset('./dataset2.csv')
```

```python
x = [point[0] for point in dataset]
y = [point[1] for point in dataset]
```

```python
 y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(dataset)
# y_pred = AgglomerativeClustering(n_clusters=2).fit_predict(dataset)
# y_pred = DBSCAN().fit_predict(dataset)
```

```python
plt.scatter(x, y, c=y_pred)
plt.title("Kmeans")
```

```python
y_pred = DBSCAN().fit_predict(dataset)
```

```python
plt.scatter(x, y, c=y_pred)
plt.title("DBSCAN")
```

```python
y_pred = AgglomerativeClustering(n_clusters=2).fit_predict(dataset)
```

```python
plt.scatter(x, y, c=y_pred)
plt.title("AgglomerativeClustering")
```
